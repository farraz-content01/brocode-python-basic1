### Feat 1: Scientific Calculator

```python
# Create a Scientific Calculator

import math

while True:
    #! --- calculator code - start here ---

    # 1.display menu
    print("Calculator Menu:")
    print("1. Square root")
    print("2. Sine")
    print("3. Cosine")
    print("4. Tangent")
    print("5. Logarithm")
    print("6. Factorial")
    print("7. Exit")

    # 2.user input - choose operator
    operator = int(input("Enter your choice (1-7): "))

    if operator == 7:
        print("=> Goodbye!")
        break    #TODO: exit the loop

    # 3.user input - get data (numbers)
    try:
        num = float(input("Enter the number: "))
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
        continue    #TODO: back to step 2

    # 4. perform the requested operation
    if operator == 1:
        opr = "sqrt"
        result = math.sqrt(num)
    elif operator == 2:
        opr = "sin"
        result = math.sin(num)
    elif operator == 3:
        opr = "cos"
        result = math.cos(num)
    elif operator == 4:
        opr = "tan"
        result = math.tan(num)
    elif operator == 5:
        opr = "log"
        result = math.log(num)
    elif operator == 6:
        opr = "factorial"
        result = math.factorial(int(num))

    else:
        print("Invalid choice. Please try again.")
        continue    #TODO: back to step 2

    # 5. display the result
    print("# Selected operator:", opr)
    print(f"Result: {result}")

    #! --- calculator code - end here ---
    print()

    # 6. ask user if they want to continue
    continue_calc = input("Do you want to continue? (Y/N): ")
    if continue_calc.lower() != "y":   # if user input is not 'Y'
        print("Goodbye!")
        break    #TODO: exit the loop 
```


### Feat 2: Calculator with History

```python
# Calculator with History
# Add a feature to store and display calculation history

history = []

while True:
    #! --- calculator code - start here ---

    # 1.display menu
    print("Calculator Menu:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    # 2.user input - choose operator
    choice = int(input("Enter your choice (1-5): "))

    if choice == 5:
        print("=> Goodbye!")
        break    #TODO: exit the loop

    # 3.user input - get data (numbers)
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
        continue    #TODO: back to step 2

    # 4. perform the requested operation
    if choice == 1:
        opr = "+"
        result = num1 + num2
    elif choice == 2:
        opr = "-"
        result = num1 - num2
    elif choice == 3:
        opr = "*"
        result = num1 * num2
    elif choice == 4:
        opr = "/"
        result = num1 / num2
    else:
        print("Invalid choice. Please try again.")
        continue    #TODO: back to step 2

    # 5. display the result
    print(f"Result: {num1} {opr} {num2} = {result}")

    # 6. add the calculation to history
    history.append(f"{num1} {opr} {num2} = {result}")

    # 7. display the history
    print("Calculation History:")
    for i, calc in enumerate(history, start=1):
        print(f"{i}. {calc}")

    #! --- calculator code - end here ---
    print()

    # 8. ask user if they want to continue
    continue_calc = input("Do you want to continue? (Y/N): ")
    if continue_calc.lower() != "y":   # if user input is not 'Y'
        print("Goodbye!")
        break    #TODO: exit the loop

```
