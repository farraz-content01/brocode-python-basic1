### Feat 1: Scientific Calculator

```python
# Create a Scientific Calculator

import math

while True:
    #! --- calculator code - start here ---

    # 1.display menu
    print("Calculator Menu:")
    print("1. Square root")
    print("2. Sine")
    print("3. Cosine")
    print("4. Tangent")
    print("5. Logarithm")
    print("6. Factorial")
    print("7. Exit")

    # 2.user input - choose operator
    operator = int(input("Enter your choice (1-7): "))

    if operator == 7:
        print("=> Goodbye!")
        break    #TODO: exit the loop

    # 3.user input - get data (numbers) > with error handling
    try:
        num = float(input("Enter the number: "))
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
        continue    #TODO: back to step 2

    # 4. perform the requested operation 
    # --scientific calculations using math module
    if operator == 1:
        opr = "sqrt"
        result = math.sqrt(num)
    elif operator == 2:
        opr = "sin"
        result = math.sin(num)
    elif operator == 3:
        opr = "cos"
        result = math.cos(num)
    elif operator == 4:
        opr = "tan"
        result = math.tan(num)
    elif operator == 5:
        opr = "log"
        result = math.log(num)
    elif operator == 6:
        opr = "factorial"
        result = math.factorial(int(num))
    else:
        print("Invalid choice. Please try again.")
        continue    #TODO: back to step 2

    # 5. display the result
    print("# Selected operator:", opr)
    print(f"- Result: {result}")

    #! --- calculator code - end here ---
    print()

    # 6. ask user if they want to continue
    continue_calc = input("Do you want to continue? (Y/N): ")
    if continue_calc.lower() != "y":   # if user input is not 'Y'
        print("Goodbye!")
        break    #TODO: exit the loop 
```


### Feat 2: Calculator with History

```python
# Calculator with History
# Add a feature to store and display calculation history

history = []

while True:
    #! --- calculator code - start here ---

    # 1.display menu
    print("Calculator Menu:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    # 2.user input - choose operator
    choice = int(input("Enter your choice (1-5): "))

    if choice == 5:
        print("=> Goodbye!")
        break    #TODO: exit the loop

    # 3.user input - get data (numbers) > with error handling
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
        continue    #TODO: back to step 2

    # 4. perform the requested operation
    if choice == 1:
        opr = "+"
        result = num1 + num2
    elif choice == 2:
        opr = "-"
        result = num1 - num2
    elif choice == 3:
        opr = "*"
        result = num1 * num2
    elif choice == 4:
        opr = "/"
        result = num1 / num2
    else:
        print("Invalid choice. Please try again.")
        continue    #TODO: back to step 2

    # 5. display the result 
    print("# Selected operator:", opr)
    print(f"Result: {num1} {opr} {num2} = {result}")

    # 6. add the calculation to history  
    history.append(f"{num1} {opr} {num2} = {result}")
    # - use a 'list' to store history > use append() method
    # - append() method is used to add an item to the end of the list


    # 7. display the history 
    print("# Calculation History:")
    for i, calc in enumerate(history, start=1):
        print(f"{i}. {calc}")
    
    # EX: Calculation History:
    # 1. 223.0 + 66.0 = 289.0
    # 2. 89.0 / 9.0 = 9.88888888888889

    #! --- calculator code - end here ---
    print()

    # 8. ask user if they want to continue
    continue_calc = input("Do you want to continue? (Y/N): ")
    if continue_calc.lower() != "y":   # if user input is not 'Y'
        print("Goodbye!")
        break    #TODO: exit the loop

   # 9. add a feature to clear history
    clear_history = input("Do you want to clear history? (Y/N): ")
    if clear_history.lower() == "y":   # if user input is 'Y'
        history.clear()    #TODO: clear the history list
        print("History cleared.")
    print()

    # 10 . add a feature to save history to a file
    save_history = input("Do you want to save history to a file? (Y/N): ")
    if save_history.lower() == "y":   # if user input is 'Y'
        with open("calculation_history.txt", "w") as file:   #TODO: open a file in write mode
            for calc in history:
                file.write(calc + "\n")    #TODO: write each calculation to the file
        print("History saved to calculation_history.txt")
    print()

    # 11. add a feature to load history from a file
    load_history = input("Do you want to load history from a file? (Y/N): ")
    if load_history.lower() == "y":   # if user input is 'Y'
        try:
            with open("calculation_history.txt", "r") as file:   #TODO: open a file in read mode
                history = [line.strip() for line in file]    #TODO: read each line and store in history list
            print("History loaded from calculation_history.txt")
        except FileNotFoundError:
            print("No history file found.")
    print()

    # 12. add a feature to clear the console
    clear_console = input("Do you want to clear the console? (Y/N): ")
    if clear_console.lower() == "y":   # if user input is 'Y'
        import os
        os.system('cls' if os.name == 'nt' else 'clear')    #TODO: clear the console
    print()

    # 13. add a feature to exit the program
    exit_program = input("Do you want to exit the program? (Y/N): ")
    if exit_program.lower() == "y":   # if user input is 'Y'
        print("Goodbye!")
        break    #TODO: exit the loop

```
