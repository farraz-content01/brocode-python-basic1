### Challenge 1: Multi-Step Calculator

```python
# Create a multi-step calculator program with the following features:
# 1. display a menu of operations (add, subtract, multiply, divide)
# 2. ask the user to select an first operation
# 3. user input - get data (numbers) > with error handling
# 4. perform the calculation
# 5. ask the user to select an second operation (optional) > if the user selects an operation, repeat steps 3-4 with the result of the previous calculation as the first number 
# 6. add the calculation to history
# 7. display the history
# 8. allow the user to exit the program

# 1. display a menu of operations (add, subtract, multiply, divide)
print("== Welcome to the Calculator! ==")
print("Select the 1st operation:")
print("1. Add (+)")
print("2. Subtract (-)")
print("3. Multiply (*)")
print("4. Divide (/)")
print("5. Exit")

# 2. ask the user to select an first operation
history = []   # to store calculation history
while True:
    # -- first operation --
    opr1 = input("Enter the number corresponding to the operation (1-5): ")
    
    if opr1 == '5':
        print("Exiting the calculator. Goodbye!")
        break
    
    if opr1 not in ['1', '2', '3', '4']:
        print("Invalid selection. Please try again.")
        continue
    
    # 3. user input - get data (numbers) > with error handling
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        continue
    
    # 4. perform the calculation
    if opr1 == '1':
        result1 = num1 + num2
    elif opr1 == '2':
        result1 = num1 - num2
    elif opr1 == '3':
        result1 = num1 * num2
    elif opr1 == '4':
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
            continue
        result1 = num1 / num2

    # Display the selected 1st operator - using a 'dictionary' to map the operator number to its symbol 
    print(f"# Selected operator: {'+' if opr1 == '1' else '-' if opr1 == '2' else '*' if opr1 == '3' else '/'}")
    # Display the result of the first operation
    print(f"--Result (1st operation): {result1}")

    # 5. ask the user to select an second operation (optional) 
    # - if the user selects an operation, repeat steps 3-4 
    # - with the result of the previous calculation as the first number
    print("Select the 2nd operation (optional):")
    print("1. Add (+)")
    print("2. Subtract (-)")
    print("3. Multiply (*)")
    print("4. Divide (/)")
    print("5. No further operation")
    
    # -- second operation --
    opr2 = input("Enter the number corresponding to the operation (1-5): ")

    if opr2 == '5':
        result2 = result1
    elif opr2 in ['1', '2', '3', '4']:
        try:
            num3 = float(input("Enter the second number for the 2nd operation: "))
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
            continue
        
        if opr2 == '1':
            result2 = result1 + num3
        elif opr2 == '2':
            result2 = result1 - num3
        elif opr2 == '3':
            result2 = result1 * num3
        elif opr2 == '4':
            if num3 == 0:
                print("Error: Division by zero is not allowed.")
                continue
            result2 = result1 / num3
    else:
        print("Invalid selection for the 2nd operation. Please try again.")
        continue

    # Display the final result of the calculation
    if opr2 == '5':
        final_result = result1  # No further operation, use the result of the first operation as the final result
    else:
        final_result = result2  # Use the result of the second operation as the final result    
    
    # Display the selected 2nd operator
    opr_symbol = {'1': '+', '2': '-', '3': '*', '4': '/'}
    if opr2 in ['1', '2', '3', '4']:
        print(f"# Selected operator for 2nd operation: {opr_symbol[opr2]}")
    print(f"--Final Result: {final_result}")

    # 6. add the calculation to history
    if opr2 == '5':
        history.append(f"{num1} {opr_symbol[opr1]} {num2} = {final_result}")
    else:
        history.append(f"{num1} {opr_symbol[opr1]} {num2} {opr_symbol[opr2]} {num3} = {final_result}") 

    # 7. display the history
    print("# Calculation History:")
    for i, calc in enumerate(history, start=1):
        print(f"{i}. {calc}")   
    print()

    # 8. ask the user if they want to run the calculator again
    run_again = input("Do you want to run the calculator again? (yes/no): ")
    if run_again.lower() != "yes":
        print("Exiting the calculator. Goodbye!")
        break

print("Thank you for using the calculator. Goodbye!")
```
